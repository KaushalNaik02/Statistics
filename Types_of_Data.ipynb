Types of Data

## Qualitative Data

```python
import pandas as pd
import numpy as np
np.random.seed(0)
```

#### Nominal Data

```python
num_sample = 1000
data = ['red','green','yellow','blue','black','brown']
dataarray = np.random.choice(data,num_sample)
dataarray
final = pd.DataFrame(dataarray,columns=['color'])
final
```

### Ordinal Data

```python
num_sample = 1000
data1 = ['bad','good','better']
dataarray1 = np.random.choice(data1,num_sample)
dataarray1
final['status']= dataarray1
final.head()
```

## Quantitive Data

### Internal Data ( Non True Zero)

```python
np.random.seed(0)
num_sample = 1000
temp = np.random.randint(0,100,num_sample)
final['temperature'] = temp
final.head()
```

```python
np.random.seed(0)
num_sample = 1000
sal = np.random.randint(10000,100000,num_sample)
final['income'] = sal
final.head()
```

## Plotting Distribution Graph for income

```python
import matplotlib.pyplot as plt

x = final['income']
plt.hist(x, bins=20, edgecolor='black')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.show()
```
